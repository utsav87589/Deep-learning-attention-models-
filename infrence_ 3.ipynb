```python
french_sentences = [f"<sos> {s} <eos>" for s in fra_texts]

# Initialize tokenizer
fra_tok = Tokenizer(filters="", lower=True, oov_token="<unk>")
fra_tok.fit_on_texts(french_sentences)

# Convert sentences to sequences
fra_sequences = fra_tok.texts_to_sequences(french_sentences)

# Vocabulary size
fra_vocabs = len(fra_tok.word_index) + 1  # +1 for padding

# Reverse dictionary
reverse_fra_tokens = {v: k for k, v in fra_tok.word_index.items()}

# Pad sequences
max_fra_word = max(len(seq) for seq in fra_sequences)
decoder_input_data = pad_sequences(fra_sequences, maxlen=max_fra_word, padding="pre")
```

### vocabs

```python
eng_vocabs = len(eng_tok.word_index) + 1
fra_vocabs = len(fra_tok.word_index) + 1
```

```python
print(f"{eng_vocabs} :: {fra_vocabs}")
```

### making the encoders

```python
from tensorflow.keras.layers import Input, Embedding, LSTM
```

```python
latent_dim = 256
```

```python
enc_inputs = Input(shape = (max_eng_word, ))
enc_emb = Embedding(eng_vocabs, latent_dim)(enc_inputs)
enc_lstm = LSTM(latent_dim, return_state = True, return_sequences = True)
enc_outputs, state_h, state_c = enc_lstm(enc_emb)
```

### making the decoders

```python
dec_inputs = Input(shape = (max_fra_word))
dec_emb_layer = Embedding(fra_vocabs, latent_dim)
dec_emb = dec_emb_layer(dec_inputs)
dec_lstm = LSTM(latent_dim, return_state = True, return_sequences = True)
```

### building the final Bahdanau attention model

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Layer, Concatenate, Dense
```

```python
class BahdanauAttention(Layer) : 
    def __init__(self, units) :
        super(BahdanauAttention, self).__init__()
        self.W1 = Dense(units)
        self.W2 = Dense(units)
        self.V = Dense(1)

    def call(self, query, values) :
        query_wuth_time_axis = tensorflow.expand_dims(query, 1)
        score = self.V(tensorflow.nn.tanh(self.W1(values)) + self.W2(query_wuth_time_axis))
        attention_weights = tensorflow.nn.softmax(score, 1)
        context_vector = attention_weights * score
        context_vector = tensorflow.reduce_sum(context_vector, 1)

        return context_vector, attention_weights
```

```python
attention = BahdanauAttention(latent_dim)
all_outputs = []
decoder_state_h, decoder_state_c = state_h, state_c
for t in range(max_fra_word) : 
    context_vector, _ = attention(decoder_state_h, decoder_state_c)
    X = tensorflow.expand_dims(dec_emb[:, t], 1)
    X = Concatenate(axis = -1)([context_vector[:, None, :], X])
    outputs, decoder_state_h, decoder_state_c = dec_lstm(X, initial_state = [decoder_state_h, decoder_state_c])
    outputs = Dense(fra_vocabs, activation = 'softmax')(outputs)
    all_outputs.append(outputs)

dec_outputs = tensorflow.concat(all_outputs, 1)
```

### main model and training

```python
model = Model()
```

```python
model = Model([enc_inputs, dec_inputs], dec_outputs)
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(
    [encoder_input_data, decoder_input_data],
    np.expand_dims(decoder_target_data, -1),
    batch_size = 128,
    epochs = 1,
    validation_split = 0.2
)
```

### encoders

```python
enc_model = Model(enc_inputs, [enc_outputs, state_h, state_c])
```

### decoders

```python
decoder_state_input_h = Input(shape = (latent_dim, ))
decoder_state_input_c = Input(shape = (latent_dim, ))
encoder_outputs_inputs = Input(shape = (max_eng_word, latent_dim))
decoder_input_single = Input(shape = (1, ))
decoder_emb_2 = dec_emb_layer(decoder_input_single)
context_vector, attention_weights = attention(decoder_state_input_h, encoder_outputs_inputs)
X = Concatenate(axis = -1)([tensorflow.expand_dims(context_vector, 1), decoder_emb_2])
decoder_outputs, state_h_inf, state_c_inf = dec_lstm(X, initial_state = [decoder_state_input_h, decoder_state_input_c])
decoder_outputs = Dense(fra_vocabs, activation = 'softmax')(decoder_outputs)

dec_model = Model(
    [decoder_input_single, decoder_state_input_h, decoder_state_input_c, encoder_outputs_inputs],
    [decoder_outputs, state_h_inf, state_c_inf, attention_weights]
)
```

### final decode sequence function

```python
def decode_sequence(input_seq) : 
    enc_outs, state_h, state_c = enc_model.predict(input_seq)

    target_seq = np.array([[fra_tok.word_index['<sos>']]])
    
    stop_condition = False
    decode_sentence = []

    while not stop_condition :
        output_tokens, h, c, attn = dec_model.predict([target_seq, state_h, state_c, enc_outs])

        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_word = reverse_fra_tokens.get(sampled_token_index, '')

        if (sampled_word == '<eos>' or len(decode_sentence) > max_fra_word) :
            stop_condition = True
        else :
            decode_sentence.append(sampled_word)

        target_seq = np.array([[sampled_token_index]])

        state_h, state_c = h, c

        print(sampled_token_index, sampled_word)


    return ' '.join(decode_sentence)
```
