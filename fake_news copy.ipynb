## Fake news classifier

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Final.csv')
df.head()
```

```python
df = df.sample(frac = 1).reset_index(drop = True)
df.head()
```

```python
df.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
df.head()
```

```python
X = df.drop('info', axis = 1)
y = df['info']
```

### text preprocessing

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
nltk.download('stopwords')
```

```python
messages = X.copy()
```

```python
corpus = []
ps = PorterStemmer()
for i in range(0 ,len(messages)) : 
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])

    review = review.lower()
    review = review.split()

    review = [ps.stem(words) for words in review if not words in stopwords.words('english')]
    review = ' '.join(review)

    corpus.append(review)
```

```python
corpus
```

### word embedding

```python
import tensorflow
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
max_len = 20
vocab_size = 10000
```

```python
one_hot_rep = [one_hot(words, vocab_size) for words in corpus]
one_hot_rep
```

```python
word_embedded = pad_sequences(one_hot_rep, maxlen = max_len, padding = 'post')
word_embedded
```

```python
X_final = np.array(word_embedded)
y_final = np.array(y)
```

```python
X_final
```

```python
y_final
```
