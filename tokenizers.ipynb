## Tokenizers

### Data loading

```python
import pandas as pd
import numpy as np
```

```python
path = 'fra.txt'
eng_texts, fra_texts = [], []
num_samples = 10000
```

```python
with open(path, 'r', encoding = 'UTF-8') as f :
    lines = f.read().split('\n')

    for line in lines[:num_samples] : 
        eng_words, fra_words = line.split('\t')[:2]

        eng_texts.append(eng_words)
        fra_texts.append(fra_words)
```

### tokenizers

### english tokenizers

```python
import tensorflow
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
eng_tokenizer = Tokenizer(filters = ' ', lower = True)
eng_tokenizer.fit_on_texts(eng_texts)
eng_sequence = eng_tokenizer.texts_to_sequences(eng_texts)
```

### french tokenizers

```python
fra_texts = []
```

```python
french_sentences = [f"<sos> {s} <eos>" for s in fra_texts]

# Initialize tokenizer
fra_tok = Tokenizer(filters="", lower=True, oov_token="<unk>")
fra_tok.fit_on_texts(french_sentences)

# Convert sentences to sequences
fra_sequences = fra_tok.texts_to_sequences(french_sentences)

# Vocabulary size
fra_vocabs = len(fra_tok.word_index) + 1  # +1 for padding

# Reverse dictionary
reverse_fra_tokens = {v: k for k, v in fra_tok.word_index.items()}

# Pad sequences
max_fra_word = max(len(seq) for seq in fra_sequences)
fra_padded = pad_sequences(fra_sequences, maxlen=max_fra_word, padding="pre")

print("Word index:", fra_tok.word_index)
print("Reverse dict:", reverse_fra_tokens)
print("Padded Seqs:", fra_padded)
print("Vocab Size:", fra_vocabs)
print("Max French length:", max_fra_word)
```
